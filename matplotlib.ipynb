กราฟวงกลม

```python
student=[10,20,30,40,50]
course=['PHP','Java','Python','C#','Nodejs']
```

```python
plt.pie(student,labels=course,)
plt.show()
```

```python
plt.pie(student,labels=course,autopct='%.1f%%')#ทศนิยม1ตำแหน่ง
plt.show()
```

ปรับแต่งกราฟวงกลม

```python
color=['purple','red','cyan','yellow','green']
exp=[0,1,0,0.2,0]#เอาชิ้นส่วนในวงกลมออกมา
plt.pie(student,labels=course,autopct='%.1f%%',colors=color,shadow=True,explode=exp)#ทศนิยม1ตำแหน่ง
plt.show()
```

```python
color=['purple','red','cyan','yellow','green']
exp=[0,1,0,0.2,0]#เอาชิ้นส่วนในวงกลมออกมา
plt.pie(student,labels=course,autopct='%.1f%%',colors=color,shadow=True,explode=exp,startangle=45)#ทศนิยม1ตำแหน่ง
plt.show()
```

กราฟแท่ง

```python
course=['python','java','C#']
score=[80,75,50]
plt.bar(course,score)
plt.xlabel('course')
plt.ylabel('score')
plt.title('score midterm')
plt.show()
```

```python
course=['python','java','C#']
score=[80,75,50]
c=['green','red','blue']
plt.bar(course,score,color=c,width=0.6,edgecolor='black',linewidth=3)
plt.xlabel('course')
plt.ylabel('score')
plt.title('score midterm')
plt.show()
```

```python
course=['python','java','C#']
score=[80,75,50]
c=['green','red','blue']
plt.barh(course,score,color=c,edgecolor='black',linewidth=3)
plt.xlabel('score')
plt.ylabel('course')
plt.title('score midterm')
plt.show()
```

stackbar

```python
room=['a','b','c']
boys=[10,15,5]
girls=[20,7,15]
plt.bar(room,boys,label='boys',color='blue')
plt.bar(room,girls,bottom=boys,label='girls',color='red')
plt.legend()
plt.show()
```

```python
#ทำเป็นarea
room=['a','b','c']
boys=[10,15,5]
girls=[20,7,15]
plt.stackplot(room,boys,girls,labels=['boys','girls'],colors=['green','pink'])
plt.legend()
```
